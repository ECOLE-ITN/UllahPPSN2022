```python
import numpy as np
import pandas as pd
import os
import sys
import pickle5 as pickle
import Utils
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
plt.style.use('ggplot')

fixed_iterations = pd.read_csv ("Fixed Iterations\\Fixed__Iterations.csv", index_col = 0)

fixed_time = pd.read_csv ("Fixed Time\\Fixed__Time.csv", index_col = 0)

fixed_target = pd.read_csv ("Fixed Target\\Fixed__Target.csv", index_col = 0)

Avg_cpu = pd.read_csv ('Avg CPU Time\\Avg__CPUT.csv', index_col = 0)

def ecdf (data):
    
    """ 
    Compute ECDF
    
    """
    
    x = np.sort (data)
    n = x.size
    y = np.arange (1, n+1) / n
    return (x,y)
```
