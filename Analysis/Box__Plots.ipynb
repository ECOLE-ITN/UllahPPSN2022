```python
import numpy as np
import pandas as pd
import os
import sys
import pickle5 as pickle
import Utils
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
plt.style.use('ggplot')

data_2D = pd.read_csv ('2D.csv', index_col = 0) 
data_5D = pd.read_csv ('5D.csv', index_col = 0) 
data_10D = pd.read_csv ('10D.csv', index_col = 0) 
```

```python
data = pd.DataFrame (np.append (data_2D ['MMR'].values, [data_2D ['MMRR'].values,
                                                 data_2D ['EBR'].values,
                                                 data_2D ['DBR'].values,
                                                 data_2D ['CR'].values]), columns = ['CPU Time (Seconds)'])
data ['RF'] = np.append (['MMR'] * 20,[['MMRR'] * 20, ['EBR'] * 20, ['DBR'] * 20, ['CR'] * 20])

data_1 = pd.DataFrame (np.append (data_5D ['MMR'].values, [data_5D ['MMRR'].values,
                                                 data_5D ['EBR'].values,
                                                 data_5D ['DBR'].values,
                                                 data_5D ['CR'].values]), columns = ['CPU Time (Seconds)'])
data_1 ['RF'] = np.append (['MMR'] * 20,[['MMRR'] * 20, ['EBR'] * 20, ['DBR'] * 20, ['CR'] * 20])

data_2 = pd.DataFrame (np.append (data_10D ['MMR'].values, [data_10D ['MMRR'].values,
                                                 data_10D ['EBR'].values,
                                                 data_10D ['DBR'].values,
                                                 data_10D ['CR'].values]), columns = ['CPU Time (Seconds)'])
data_2 ['RF'] = np.append (['MMR'] * 20,[['MMRR'] * 20, ['EBR'] * 20, ['DBR'] * 20, ['CR'] * 20])
```

```python
fig, axs = plt.subplots(1, 3, figsize = (12, 6))

axs[0] = sns.boxplot (data = data, y = 'CPU Time (Seconds)', x = 'RF', palette = 
                      ['r', 'b', 'g', 'k', 'm'], saturation = 0.6, ax = axs[0], notch = True, sym="")

axs[0].set_yscale ('log', basey = 2)
axs[0].set_ylabel("$T_{\mathrm{max}}$ (Seconds)",fontsize=20)
axs[0].set (xlabel = "")
axs[0].tick_params (labelsize = 15)
axs[0].set_xticklabels(["MMR", "MMRR", "EBR", "DBR", "CR"], rotation=90)
axs[0].set_title("D = 2", size = 15)

axs[1] = sns.boxplot (data = data_1, y = 'CPU Time (Seconds)', x = 'RF', palette = 
                      ['r', 'b', 'g', 'k', 'm'], saturation = 0.6, ax = axs[1], notch = True, sym ="")

axs[1].set_yscale ('log', basey = 2)
axs[1].set (ylabel = "")
axs[1].set (xlabel = "")
axs[1].tick_params (labelsize = 15)
axs[1].set_xticklabels(["MMR", "MMRR", "EBR", "DBR", "CR"], rotation=90)
axs[1].set_title("D = 5", size = 15)

axs[2] = sns.boxplot (data = data_2, y = 'CPU Time (Seconds)', x = 'RF', palette = 
                      ['r', 'b', 'g', 'k', 'm'], saturation = 0.6, ax = axs[2], notch = True, sym = "")

axs[2].set_yscale ('log', basey = 2)
axs[2].set (ylabel = "")
axs[2].set (xlabel = "")
axs[2].tick_params (labelsize = 15)
axs[2].set_xticklabels(["MMR", "MMRR", "EBR", "DBR", "CR"], rotation=90)
axs[2].set_title("D = 10", size = 15)

fig.tight_layout()  
fig.savefig ( "Box.pdf", bbox_inches = 'tight')
```
